
# Background: 
Breast cancer is traditionally stratified using immunohistochemical subtypes, since molecular analysis is very expensive. 

# Research question: 
Can molecular information from the METABRIC dataset improve survival prediction beyond conventional immunohistological classification?

# Hypothesis:
Molecular information improves models predicting patient survival, compared to immunohistochemical and clinical data.  

# Approach
1. Create a model to predict overall survival (months) from only clinical data (control)
2. Create a model to predict overall survival (months) from clinical data and immunohistological data
3. Create a model to predict over survival (months) from clinical and molecular data
4. Compare the 3 models and see which performs better
